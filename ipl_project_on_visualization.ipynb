# **IPL DATA ANALYSIS fROM 2008 to 2016**

**The IPL (Indian Premier League) Data Visualization project uses Python and various libraries such as Pandas, NumPy, and Matplotlib to explore and visualize key insights from IPL match data. The dataset includes information about teams, cities, toss decisions, winners, and other details related to IPL matches. The goal of this project is to provide a comprehensive analysis of the data, enabling a better understanding of team performance, match statistics, and trends over the years.**

# **Data Visualization Project**

```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading IPL match data from a CSV file into a DataFrame
data = pd.read_csv('/content/data.csv')
```

```python
# Displaying basic information about the dataset
print("Dataset Information:")
print(data.info())
```

```python
# Displaying the first few rows of the dataset
print("First 5 rows of the dataset:")
print(data.head())
```

```python
# Summary statistics of numerical columns
print("Summary Statistics:")
print(data.describe())
```

# **EDA**

```python
# Team-wise statistics
team_stats = pd.concat([data['team1'], data['team2']], axis=0).value_counts()
print("Team-wise Statistics:")
print(team_stats)
```

```python

# Total matches played in each city
city_stats = data['city'].value_counts()
print("Matches Played in Each City:")
print(city_stats)
```

```python

# Matches won by each team
winning_teams = data['winner'].value_counts()
print("Matches Won by Each Team:")
print(winning_teams)
```

# **VISUALIZATION**

```python

# Plotting the distribution of matches won by each team
plt.figure(figsize=(10, 6))
winning_teams.plot(kind='barh', color='skyblue')
plt.title('Matches Won by Each Team')
plt.xlabel('Number of Matches Won')
plt.ylabel('Teams')
plt.show()
```

```python
# Distribution of matches played in each city
plt.figure(figsize=(12, 6))
plt.bar(city_stats.index, city_stats.values, color='lightgreen')
plt.title('Matches Played in Each City')
plt.xlabel('City')
plt.ylabel('Matches Played')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python

# Plotting the distribution of toss decisions
plt.figure(figsize=(8, 8))
data['toss_decision'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['gold', 'lightcoral'])
plt.title('Distribution of Toss Decisions')
plt.show()
```

```python
# Counting the number of toss wins based on decisions
toss_decision_counts = data['toss_decision'].value_counts()
print(toss_decision_counts)
# Plotting the teams with the highest and lowest toss wins based on toss decisions
plt.figure(figsize=(10, 6))
toss_decision_counts.plot(kind='bar', color=['lightcoral' if decision == toss_decision_counts.min() else 'gold' for decision in toss_decision_counts])
plt.title('Teams with the Highest and Lowest Number of Toss Wins Based on Toss Decisions')
plt.xlabel('Toss Decision')
plt.ylabel('Number of Toss Wins')
plt.xticks(rotation=0)
plt.show()
```

# **Analysis based on scenarios**

highest matches played in mumbai so finding out matches played v/s won by each team

```python
# Matches played and won in Mumbai
Mumbai_matches = data[data['city'] == 'Mumbai']
Mumbai_matches[['team1', 'team2', 'winner']]

# Visualization for Matches Played and Won in Hyderabad
plt.figure(figsize=(12, 6))
teams = pd.concat([Mumbai_matches['team1'], Mumbai_matches['team2']]).unique()

matches_played = []
matches_won = []

for team in teams:
    matches_played.append(Mumbai_matches[(Mumbai_matches['team1'] == team) | (Mumbai_matches['team2'] == team)].shape[0])
    matches_won.append(Mumbai_matches[Mumbai_matches['winner'] == team].shape[0])

# Print the matches played and won for each team
for team, played, won in zip(teams, matches_played, matches_won):
    print(f"{team}: Played - {played},Won - {won}")

bar_width = 0.35
index = range(len(teams))
plt.bar(index, matches_played, width=bar_width, label='Matches Played', color='skyblue')
plt.bar([i + bar_width for i in index], matches_won, width=bar_width, label='Matches Won', color='lightcoral')

plt.title('Matches Played and Won in Mumbai')
plt.xlabel('Teams')
plt.ylabel('Number of Matches')
plt.xticks([i + bar_width/2 for i in index], teams, rotation=45, ha='right')
plt.legend()
plt.show()
```

lowest matches played in kanpur so finding out matches played v/s win by each team

```python
# Matches played and won in Kanpur
Kanpur_matches = data[data['city'] == 'Kanpur']
Kanpur_matches[['team1', 'team2', 'winner']]

# Visualization for Matches Played and Won in Kanpur
plt.figure(figsize=(12, 6))

# Extracting unique teams in Kanpur matches
teams = pd.concat([Kanpur_matches['team1'], Kanpur_matches['team2']]).unique()


# Lists to store the number of matches played and won for each team
matches_played = []
matches_won = []

# Calculating the number of matches played and won for each team
for team in teams:
    matches_played.append(Kanpur_matches[(Kanpur_matches['team1'] == team) | (Kanpur_matches['team2'] == team)].shape[0])
    matches_won.append(Kanpur_matches[Kanpur_matches['winner'] == team].shape[0])


# Print the matches played and won for each team
for team, played, won in zip(teams, matches_played, matches_won):
    print(f"{team}: Played - {played},Won - {won}")

# Bar chart settings
bar_width = 0.10
index = range(len(teams))
plt.bar(index, matches_played, width=bar_width, label='Matches Played', color='skyblue')
plt.bar([i + bar_width for i in index], matches_won, width=bar_width, label='Matches Won', color='lightcoral')

# Chart labels and title
plt.title('Matches Played and Won in Kanpur')
plt.xlabel('Teams')
plt.ylabel('Number of Matches')

# Customizing x-axis ticks to show team names
plt.xticks([i + bar_width/2 for i in index], teams, rotation=45, ha='right')

# Adding legend
plt.legend()

# Displaying the plot
plt.show()
```

```python
# Loading IPL match data from a CSV file into a DataFrame
data = pd.read_csv('/content/data.csv')

# Matches won after 2010
matches_after_2010 = data[data['date'] > '2010-01-01']
matches_after_2010[['team1', 'team2', 'winner', 'date']]

# Convert 'winner' column to string type
matches_after_2010['winner'] = matches_after_2010['winner'].astype(str)

# Counting matches won by each team
winning_teams_count = matches_after_2010['winner'].value_counts()

# Visualization for Matches Won After 2010 (Bar chart)
plt.figure(figsize=(12, 6))
winning_teams_count.plot(kind='bar', color='skyblue')
plt.title('Matches Won After 2010')
plt.xlabel('Teams')
plt.ylabel('Number of Matches Won')
plt.xticks(rotation=45, ha='right')
plt.show()
```
